```python
def orthogonal(x, y, z):
        if ((x @ y) == 0) and ((x @ z) == 0) and ((y @ z) == 0):
            print('Orthogonal +')
        else:
            print('Orthogonal -')
            
def vector_length(x, y, z):
    if (norm(x) == norm(y)) and (norm(x) == norm(z)) and (norm(y) == norm(z)):
        print('length +')
    else:
        print('length -')
```

```python
# A
x = np.array([1, 0, 0])
y = np.array([0, 0, 1])

if ((x @ y) == 0) and (norm(x) == norm(y)):
    print('Orthogonal +')
    print('length +')
else:
    print('Orthogonal - or length -')
```

```python
# Б
x = np.array([1/pow(2, 0.5), -1/pow(2, 0.5), 0])
y = np.array([1/pow(2, 0.5), 1/pow(2, 0.5), 0])
z = np.array([0, 0, 1])

orthogonal(x, y, z)
vector_length(x, y, z)
```

```python
# В
x = np.array([1/2, -1/2, 0])
y = np.array([0, 1/2, 1/2])
z = np.array([0, 0, 1])

orthogonal(x, y, z)
vector_length(x, y, z)
```

```python
# Г
x = np.array([1, 0, 0])
y = np.array([0, 1, 0])
z = np.array([0, 0, 1])

orthogonal(x, y, z)
vector_length(x, y, z)
```
