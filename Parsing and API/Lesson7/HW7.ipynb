Взять любую категорию товаров на сайте Ситилинк. Собрать следующие данные:    
● название;  
● все фото;  
● ссылка;  
● цена.  

```python
#только при помощи данного пакета удалось открыть ЛеройМерлин через selenium.
#все остальные, даже selenium-silenth не работал. В результате открытия была пустая страница

# https://github.com/ultrafunkamsterdam/undetected-chromedriver
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
import lxml
from lxml import html
import time
import pymongo

start = time.time()
#create connection
client = pymongo.MongoClient('mongodb://localhost:27017')
db = client.leroymerlin


domain = 'https://leroymerlin.ru'
driver = uc.Chrome()
webpages = 236
for page in range(1, webpages+1, 1):
    driver.get(f'https://leroymerlin.ru/catalogue/plitka-keramogranit-i-mozaika/?page={page}')
    time.sleep(5) #прогрузка страницы
    html_code = lxml.html.fromstring(driver.page_source, parser = lxml.html.HTMLParser(encoding='utf-8'))
    div_cards = html_code.xpath("//div[@class='phytpj4_plp largeCard']")
    _ = 0
    for div in div_cards:
        title = str(div.xpath("./a/@aria-label"))[2:-2]
        url = domain + str(div.xpath("./div[@class='c155f0re_plp cz97q1i_plp largeCard']/a/@href"))[2:-2]
        
        try:
            price = div.xpath("./div[@class='oo1t094_plp largeCard']/div/div[@data-qa='product-primary-price']/p[@class='t3y6ha_plp xc1n09g_plp p1q9hgmc_plp']/text()")[0]
        except:
            price = div.xpath("./div[@class='oo1t094_plp largeCard o1nn3tkz_plp']/div/div/div[@class='n1s0vz55_plp']/span[@class='_3rC-Ot1yr4_plp _1pNwL6sJc8_plp nfh3x0v_plp']/text()")[0]
        
        url_image = div.xpath("//img[@class='p1g8n69v_plp']/@src")[_]
        _ += 1
        
        db.plitka.insert_one({ 
            'title': title,
            'url': url,
            'url_image': url_image,
            'price': price,
        })
    time.sleep(10)
    print(f'Page - {page}. Complete')
    
    
driver.quit()
end = time.time() - start
print(f'Lead time {end}')
```
