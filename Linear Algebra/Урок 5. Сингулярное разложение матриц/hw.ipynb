```python
import numpy as np
```

```python
A = np.array([[1, 2, 0],
              [0, 0, 5],
              [3, -4, 2],
              [1, 6, 5],
              [0, 1, 0]])
```

```python
U, s, W = np.linalg.svd(A)

V = W.T

D = np.zeros_like(A, dtype=float)
D[np.diag_indices(min(A.shape))] = s
```

```python
print(f'Матрица D:\n{D}')
```

```python
print(f'Матрица U:\n{U}')
```

```python
 print(f'Матрица V:\n{V}')
```

```python
B = np.array(np.dot(np.dot(U, D), V.T), dtype='float16')
B
```
