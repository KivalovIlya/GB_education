```python
from math import factorial
from math import e

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

#расстановки/размещения (порядок важен)
def A(n, m):
  try:
    return factorial(n) / factorial(n-m)
  except ValueError:
    print('Ошибка: m должен быть < n')


#сочетания (порядок не важен)
def C(n, m):
  try:
    return factorial(n) / (factorial(m) * factorial(n-m))
  except ValueError:
    print('Ошибка: m должен быть < n')

#формула Бернулли
def Bernoulli(p, n, k):
  Cnk = C(n, k)
  p_pow_k = pow(p, k)
  q_pow_n_k = pow(1-p, n-k)
  return Cnk*p_pow_k*q_pow_n_k

#распределение Пуассона
def Poisson_distr(p, n, m):
  hm = pow(n*p, m)
  pm = (hm/factorial(m))*pow(e, n*p*(-1))
  return pm
```

## 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
#построим график и возьмем вероятность при 85
figure(figsize=(5, 5), dpi=100)
plt.grid()
plt.plot([i for i in range(1, 101)], [Bernoulli(0.8, 100, i) for i in range(1, 101)])
plt.plot(85, Bernoulli(0.8, 100, 85), 'ro', color='red')
plt.title(str(Bernoulli(0.8, 100, 85)) + ' ответ' )
plt.show()
```
